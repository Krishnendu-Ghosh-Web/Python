```python
#Working with Images with Python
from PIL import Image
krish=Image.open("boat.jpg")
type(krish) 
krish
print(krish.size)
print(krish.filename)
print(krish.format_description)
krish
```

```python
#cropping image
krish.crop((0,0,100,100))
```

```python
noise=Image.open('noise.jpg')
noise
x=38
y=4
w=148
h=112
logo=noise.crop((x,y,w,h))
logo
```

```python
noise.paste(im=logo,box=(25,0))
noise
```

```python
noise.size
noise.resize((200,150))
noise.rotate(45)
```

```python
#RGBA (color transparency)
boult=Image.open("boult.jpg")
boult
#boult.putalpha(0)  #transparent
#boult.putalpha(900) # brighting
boult.putalpha(125) # low brighting
boult
```

```python
boult.paste(im=boult,box=(0,0),mask=boult)
boult
boult.save("krish.png")
krish=Image.open("krish.png")
krish
```
